```python
df['returns'] = np.log(df['close']).diff()
df = df.dropna()
```

```python
# Fit the GARCH model
model = arch_model(df['returns'], vol='Garch', p=1, q=1)
results = model.fit()
```

```python
# Printing the model summary
print(results.summary())
```

```python
# Forecastng volatility
forecasts = results.forecast(horizon=5)  # Forecast 5 days ahead
print(forecasts.variance.tail())
```

```python

# Extracting the conditional volatility forecasts
volatility_forecast = forecasts.variance.iloc[-1]

# Printing the volatility forecast
print(volatility_forecast)
```
